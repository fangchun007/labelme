```python
import os
import json

from PIL import Image
import numpy as np

import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
path_to_image = "4-aerial.jpg"
path_to_json = "4-aerial.json"
```

```python
image = Image.open(path_to_image)

image_np = np.asarray(image)
print(image_np.shape)

with open(path_to_json, "r") as f:
    annotation_data = json.load(f)
print("annotation_data.keys() = {}".format(annotation_data.keys()))
print("annotation_data['shapes'][0] = \n{}".format(annotation_data['shapes'][0]))
```

```python
plt.figure(figsize=(14, 8))
plt.imshow(image_np)
```

```python
################################################################
# Let's just copy paste some data and do a quick verification. #
################################################################

# Reading an image in default
# mode
image = cv2.imread(path_to_image)

# Polygon corner points coordinates
points = np.array([[380.8139534883721, 65.11627906976744], 
                [419.1860465116279, 93.6046511627907], 
                [472.6744186046512, 131.97674418604652], 
                [537.7906976744187, 180.2325581395349], 
                [587.2093023255813, 212.7906976744186], 
                [630.232558139535, 244.18604651162792], 
                [685.4651162790698, 291.27906976744185], 
                [730.8139534883721, 336.04651162790697], 
                [788.953488372093, 402.90697674418607], 
                [835.1437699680512, 464.21725239616615], 
                [876.038338658147, 521.0862619808307], 
                [914.0845070422536, 576.4084507042254], 
                [965.1408450704225, 650.7042253521128], 
                [996.485623003195, 683.0670926517572], 
                [1033.8658146964856, 712.4600638977636], 
                [1071.2460063897763, 738.6581469648562], 
                [1103.8333333333335, 758.9444444444446]]).astype(np.int32)
  
points = points.reshape((-1, 1, 2))
  
isClosed = False

# Blue color in BGR
color = (255, 0, 0)
  
# Line thickness of 2 px
thickness = 52
  
# Using cv2.polylines() method draw a Blue polygon with thickness of 1 px
image = cv2.polylines(image, [points], isClosed, color, thickness)
```

```python
plt.figure(figsize=(14, 8))
plt.imshow(image)
```
